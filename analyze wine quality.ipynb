# Import Data


````python
#read data
import pandas as pd
import numpy as np

red_df=pd.read_csv('winequality-red.csv',sep=';')
wh_df=pd.read_csv('winequality-white.csv',sep=';')
red_df.head()
wh_df.head()
````

# Explore data

````python
#explore data
red_df.info()
wh_df.info()
sum(wh_df.duplicated())
red_df.quality.nunique()
wh_df.quality.nunique()
red_df.density.mean()
red_df.columns
````

````python
#add color coloumn for the two data sets
# wh=["white"]*len(white_wine_df)
# wh_df['type']=wh
# wh_df
red_color=np.repeat('red',len(red_df))
white_color=np.repeat('white',len(wh_df))
red_df['color']=red_color
wh_df['color']=white_color
wh_df
````

# Merge datasets


````python
#merge datasets
wine_df=red_df.append(wh_df,ignore_index=True)
wine_df.shape
wine_df.to_csv("winequality_edited.csv",index=False)
wine_df
````

<h3 style="color:blue">Is a certain type of wine associated with higher quality?</h3>


````python
wine_df.groupby('color')['quality'].mean()
````


<h3 style="color:blue">What level of acidity receives the highest average rating?</h3>


````python
# View the min, 25%, 50%, 75%, max pH values with Pandas describe
wine_df.describe()
# Bin edges that will be used to "cut" the data into groups
bins=[2.720000,3.110000,3.210000,3.320000,4.010000]
# Labels for the four acidity level groups
bin_labels=['High','Moderately_High','Medium','Low']
# Creates acidity_levels column
wine_df['acidity_levels']=pd.cut(wine_df['pH'],bins,labels=bin_labels)
# Checks for successful creation of this column
wine_df.head()
# Find the mean quality of each acidity level with groupby
wine_df.groupby('acidity_levels')['quality'].mean()
````

<h3 style="color:blue">Do wines with higher alcoholic content receive better ratings?</h3>

````python
# View the median of alcohol values with Pandas describe
wine_df.describe()
# Bin edges that will be used to "cut" the data into groups
bins=[8.00,10.30,14.90]
bin_labels=['low','high']
wine_df['alcohol_levels']=pd.cut(wine_df['alcohol'],bins,labels=bin_labels)
wine_df.head()
# vall33=pd.qcut(wine_df['alcohol'], 2,labels=['low','high'],retbins=True)
# Find the mean quality of each acidity level with groupby
wine_df.groupby('alcohol_levels').quality.mean()
````

# Visualize some results

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

colors=['red','white']
color_qulaity = wine_df.groupby('color')['quality'].mean()
color_qulaity.plot(kind='bar',title="Average quality for different wines",color=colors);
plt.xlabel('colors',fontsize=18)
plt.ylabel('quality',fontsize=18);


````

<h3 style="color:blue">Plotting with matplotlib</h3>


### Average Quality Ratings by Alcohol Content

````python
# Use query to select each group and get its mean quality
median = wine_df['alcohol'].median()
low = wine_df.query('alcohol < {}'.format(median))
high = wine_df.query('alcohol >= {}'.format(median))

mean_quality_low = low['quality'].mean()
mean_quality_high = high['quality'].mean()
# Create a bar chart with proper labels

locations = [1, 2]
heights = [mean_quality_low, mean_quality_high]
labels = ['Low', 'High']
plt.bar(locations, heights, tick_label=labels)
plt.title('Average Quality Ratings by Alcohol Content')
plt.xlabel('Alcohol Content')
plt.ylabel('Average Quality Rating');
````

### Average Quality Ratings by suger Content

````python
wine_df.columns=wine_df.columns.str.replace(' ','_')
median=wine_df['residual_sugar'].median()
low_suger=wine_df.query('residual_sugar < {}'.format(median))['residual_sugar'].mean()
high_suger=wine_df.query('residual_sugar >= {}'.format(median))['residual_sugar'].mean()
# Create a bar chart with proper labels
plt.bar(['low_suger','high_suger'],[low_suger,high_suger],tick_label=['low','high'])
plt.title('Average Quality Ratings by suger Content')
plt.xlabel('suger Content')
plt.ylabel('Average Quality Rating');
````

<h3 style="color:blue">Plotting Wine Type and Quality with Matplotlib</h3>


````python
# get counts for each rating and color
color_counts = wine_df.groupby(['color', 'quality']).count()['pH']
color_counts
# get total counts for each color
color_totals = wine_df.groupby('color').count()['pH']
color_totals
# get proportions by dividing red rating counts by total # of red samples
red_proportions = color_counts['red'] / color_totals['red']
red_proportions
# get proportions by dividing white rating counts by total # of white samples
white_proportions = color_counts['white'] / color_totals['white']
white_proportions

red_proportions['9'] = 0
red_proportions

ind = np.arange(len(red_proportions))  # the x locations for the groups
width = 0.4       # the width of the bars

# plot bars
red_bars = plt.bar(ind, red_proportions, width, color='r', alpha=.7, label='Red Wine')
white_bars = plt.bar(ind + width, white_proportions, width, color='w', alpha=.7, label='White Wine')

# title and labels
plt.ylabel('Proportion')
plt.xlabel('Quality')
plt.title('Proportion by Wine Color and Quality')
locations = ind + width / 2  # xtick locations
labels = ['3', '4', '5', '6', '7', '8', '9']  # xtick labels
plt.xticks(locations, labels)

# legend
plt.legend()
````
